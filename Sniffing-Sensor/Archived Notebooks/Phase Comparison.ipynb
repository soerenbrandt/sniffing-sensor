### Graph Function

```python
import pandas as pd
import numpy as np
from scipy.signal import savgol_filter as sgf
import matplotlib.pyplot as plt
from matplotlib import style

style.use("ggplot")

%matplotlib inline

fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111)

def graph(start, end, graphs, skip=[], ax=ax):
    output = []
    num = (end-start)//3+1
    for i in map(int, np.linspace(start, end, num)):
        if i not in skip:
            df = pd.read_csv("../Scan CSV Files/Phase/{}.csv".format(i))
            df.drop('Unnamed: 0', axis=1, inplace=True)
            df = (df-df.mean())/df.std()
            output.append(list(df["0"]))
    plot_df = pd.DataFrame(output)
    plot_df = [np.mean(plot_df.loc[:,i]) for i in plot_df.columns]
    graphs.append(plot_df)
```

### Polar Compounds

```python
graphs = []
    
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111)

# IPA
graph(97, graphs)
# WATER
graph(98, graphs)
# ETHANOL
graph(99, graphs)

for i in graphs:
    ax.plot(i)

ax.set_title("Polar Compounds Comparison")
ax.legend(("IPA", "Water", "EtOH"))
ax.set_xticklabels(np.linspace(0, 12, 13))
ax.set_xticks(np.linspace(0, 240, 13))
ax.set_xlabel("Time (sec)")
ax.set_ylabel("Phase")
# plt.savefig("Important Data Images/Procedure 3 [Deep Sniff]/Polar Compounds Comparison.png")
plt.show()
```

### Experiment 2 Comparison by injection time

```python
import pandas as pd
import numpy as np
from scipy.signal import savgol_filter as sgf
import matplotlib.pyplot as plt


secs = [1, 3, 5, 10, 30, 60]

%matplotlib inline

for j in map(int, np.linspace(97, 112, 6)):
    compare = []

    for i in range(j,j+3):
        df = pd.read_csv("../Scan CSV Files/Phase/{}.csv".format(i))
        df.drop('Unnamed: 0', axis=1, inplace=True)
        df = (df-df.mean())/df.std()
        compare.append(list(df["0"]))

    fig = plt.figure(figsize=(12,12))
    ax = fig.add_subplot(1,1,1)
    
    ax.plot(np.linspace(0, len(compare[0]), len(compare[0])), compare[0],
             np.linspace(0, len(compare[1]), len(compare[1])), compare[1],
             np.linspace(0, len(compare[2]), len(compare[2])), compare[2])
    ax.set_title("{} seconds".format(secs[int((j-97)//3)]))
    
    ax.legend(("Ethanol", "IPA", "Water"))
    plt.savefig("../Scan CSV Files/Phase/Compare/exp2_{}_seconds.png".format(secs[int((j-97)//3)]))
```
